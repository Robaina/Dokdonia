# Trying different clustering methods

```python
import hdbscan
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt

res_id = 'CLUSTER_DE_GENES'
pdata = pd.read_csv(os.path.join(
    os.getcwd(),f'Results/{res_id}/Processed_Data/clust_input.tsv_processed.tsv'),
                    sep='\t', index_col='Genes')


clusterer = hdbscan.HDBSCAN(metric='l2', min_cluster_size=50)
clusterer.fit(pdata.values)
```

```python
print(f'HDBSCAN found a total of {clusterer.labels_.max() + 1} clusters')
```

```python
labels = np.unique(clusterer.labels_)
gene_ids = pdata.index
clusters = {f'C_{k}': [
    gene_id for i, gene_id in enumerate(gene_ids) if clusterer.labels_[i] == k
] 
            for k in labels}
```

```python
def plotClusters(pdata, clusters):
    n_rows = int(np.ceil(len(clusters) / 2))
    fig, axes = plt.subplots(nrows=n_rows, ncols=2)
    plt.subplots_adjust(hspace=0.3)
    coords = list(np.ndindex((n_rows, 2)))
    for n, cluster_id in enumerate(clusters):
        i, j = coords[n]
        try:
            axis = axes[i, j]
        except Exception:
            axis = axes[j]
        cluster = clusters[cluster_id]
        pdata[pdata.index.isin(cluster)].transpose().plot(
            legend=False, figsize=(15, 18), title=f'{cluster_id}, size={len(cluster)}',
            ax=axis, color='#9a9a9a', linewidth=0.8,
            marker='.', markerfacecolor='#ee9929', markersize=12)
    plt.show()

plotClusters(pdata, clusters)
```

```python
# Trying mean shift
from sklearn.cluster import MeanShift

ms = MeanShift().fit(pdata.values)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
print(f'Mean shift found a total of {len(cluster_centers)} clusters') 
```

```python
labels = np.unique(ms.labels_)
gene_ids = pdata.index
clusters = {f'C_{k}': [
    gene_id for i, gene_id in enumerate(gene_ids) if ms.labels_[i] == k
] 
            for k in labels}

plotClusters(pdata, clusters)
```

```python
# Trying K-means clustering
from sklearn.cluster import KMeans

res_id = 'CLUSTER_DE_GENES'
pdata = pd.read_csv(os.path.join(
    os.getcwd(),f'Results/{res_id}/Processed_Data/clust_input.tsv_processed.tsv'),
                    sep='\t', index_col='Genes')

km = KMeans(n_clusters=6, random_state=0).fit(pdata.values)
print(f'Found a total of {km.labels_.max() + 1} clusters')
```

```python
labels = np.unique(km.labels_)
gene_ids = pdata.index
clusters = {f'C_{k}': [
    gene_id for i, gene_id in enumerate(gene_ids) if km.labels_[i] == k
] 
            for k in labels}
```

```python
def plotClusters(pdata, clusters):
    n_rows = int(np.ceil(len(clusters) / 2))
    fig, axes = plt.subplots(nrows=n_rows, ncols=2)
    plt.subplots_adjust(hspace=0.3)
    coords = list(np.ndindex((n_rows, 2)))
    for n, cluster_id in enumerate(clusters):
        i, j = coords[n]
        try:
            axis = axes[i, j]
        except Exception:
            axis = axes[j]
        cluster = clusters[cluster_id]
        pdata[pdata.index.isin(cluster)].transpose().plot(
            legend=False, figsize=(15, 18), title=f'{cluster_id}, size={len(cluster)}',
            ax=axis, color='#9a9a9a', linewidth=0.8,
            marker='.', markerfacecolor='#ee9929', markersize=12)
    plt.show()

plotClusters(pdata, clusters)
```

# Using PATRIC pathways

First need to convert PATRIC IDs to GenBank locus tags, then search for the pathways associated to each gene

```python
import pandas as pd
import numpy as np
```

```python
patric_features = pd.read_csv('Data/Function_Annotations/PATRIC/Dokdonia_MED134_Craig_PATRIC_genome_feature.csv')
patric_pathways = pd.read_csv('Data/Function_Annotations/PATRIC/Dokdonia_MED134_PATRIC_pathways.csv')
patric_pathways_genes = pd.read_csv('Data/Function_Annotations/PATRIC/Dokdonia_MED134_Craig_PATRIC_pathways_genes.csv')
```

```python
def patricID2LocusTag(patric_id, patric_features):
    return patric_features['RefSeq Locus Tag'][patric_features['PATRIC ID'] == patric_id].item()

def locusTag2PatricID(locus_tag, patric_features):
    return patric_features['PATRIC ID'][patric_features['RefSeq Locus Tag'] == locus_tag].item()

def getPatricPathway(patric_id, patric_pathways_genes, patric_pathways):
    path_name = patric_pathways_genes['Pathway Name'][patric_pathways_genes['PATRIC ID'] == patric_id].item()
    path_class = patric_pathways['Pathway Class'][patric_pathways['Pathway Name'] == path_name].item()
    return {'name': path_name, 'class': path_class}
```
